```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.ticker import FormatStrFormatter
from scipy import stats
```

```python
df_bent_Simple = pd.read_csv("../Results/20_epoch_optimal/Bent_Simple_Corr.csv")
df_bent_Whole = pd.read_csv("../Results/20_epoch_optimal/Bent_Whole_Corr.csv")
df_bent_Blend = pd.read_csv("../Results/20_epoch_optimal/Bent_Blend_Corr.csv")

bent_data = [df_bent_Simple['BF'], df_bent_Whole['BF'], df_bent_Blend['BF']]
```

```python
df_schaf_Simple = pd.read_csv("../Results/20_epoch_optimal/Schaf_Simple_Corr.csv")
df_schaf_Whole = pd.read_csv("../Results/20_epoch_optimal/Schaf_Whole_Corr.csv")
df_schaf_Blend = pd.read_csv("../Results/20_epoch_optimal/Schaf_Blend_Corr.csv")

schaf_data = [df_schaf_Simple['BF'], df_schaf_Whole['BF'], df_schaf_Blend['BF']]
```

```python
df_kats_Simple = pd.read_csv("../Results/20_epoch_optimal/Kats_Simple_Corr.csv")
df_kats_Whole = pd.read_csv("../Results/20_epoch_optimal/Kats_Whole_Corr.csv")
df_kats_Blend = pd.read_csv("../Results/20_epoch_optimal/Kats_Blend_Corr.csv")

kats_data = [df_kats_Simple['BF'], df_kats_Whole['BF'], df_kats_Blend['BF']]
```

```python
plt.figure(dpi=400, figsize=(15, 4))

yticks = 7
medianprops = dict(color="black",linewidth=2)

sub_bent = plt.subplot(131)
plt.boxplot(bent_data, medianprops=medianprops)
sub_bent.set_ylabel('Best Fitness')
sub_bent.set_xlabel('Crossovers')
sub_bent.set_xticklabels(['Simple','Whole','Blend'])
sub_bent.set_title('Bent Cigar')
sub_bent.yaxis.set_major_locator(plt.MaxNLocator(4))
sub_bent.set_ylim(9.90,10)
sub_bent.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

sub_schaf = plt.subplot(132)
plt.boxplot(schaf_data, medianprops=medianprops)
sub_schaf.set_xlabel('Crossovers')
sub_schaf.set_xticklabels(['Simple','Whole','Blend'])
sub_schaf.set_title('Schaffers F7')
sub_schaf.yaxis.set_major_locator(plt.MaxNLocator(4))
sub_schaf.set_ylim(0,10)

sub_kats = plt.subplot(133)
plt.boxplot(kats_data, medianprops=medianprops)
sub_kats.set_xlabel('Crossovers')
sub_kats.set_xticklabels(['Simple','Whole','Blend'])
sub_kats.set_title('Katsuura')
sub_kats.yaxis.set_major_locator(plt.MaxNLocator(4))
sub_kats.set_ylim(0,10)




plt.show()
```

```python
meanBentSimple = np.mean(bent_data[0])
meanBentWhole = np.mean(bent_data[1])
meanBentBlend = np.mean(bent_data[2])

stdBentSimple = np.std(bent_data[0])
stdBentWhole = np.std(bent_data[1])
stdBentBlend = np.std(bent_data[2])

print(meanBentSimple, stdBentSimple)
print(meanBentWhole, stdBentWhole)
print(meanBentBlend, stdBentBlend)

meanSchafSimple = np.mean(schaf_data[0])
meanSchafWhole = np.mean(schaf_data[1])
meanSchafBlend = np.mean(schaf_data[2])

stdSchafSimple = np.std(schaf_data[0])
stdSchafWhole = np.std(schaf_data[1])
stdSchafBlend = np.std(schaf_data[2])

print(meanSchafSimple, stdSchafSimple)
print(meanSchafWhole, stdSchafWhole)
print(meanSchafBlend, stdSchafBlend)


meanKatsSimple = np.mean(kats_data[0])
meanKatsWhole = np.mean(kats_data[1])
meanKatsBlend = np.mean(kats_data[2])

stdKatsSimple = np.std(kats_data[0])
stdKatsWhole = np.std(kats_data[1])
stdKatsBlend = np.std(kats_data[2])

print(meanKatsSimple, stdKatsSimple)
print(meanKatsWhole, stdKatsWhole)
print(meanKatsBlend, stdKatsBlend)
```

```python
plt.figure(dpi=400, figsize=(15, 4))

df_prob_bent1 = pd.read_csv("../Results/bestProb_results/BentCigar1_prob.csv")
df_prob_bent2 = pd.read_csv("../Results/bestProb_results/BentCigar2_prob.csv")
df_prob_bent3 = pd.read_csv("../Results/bestProb_results/BentCigar3_prob.csv")

df_prob_schaf1 = pd.read_csv("../Results/bestProb_results/Schaffers1_prob.csv")
df_prob_schaf2 = pd.read_csv("../Results/bestProb_results/Schaffers2_prob.csv")
df_prob_schaf3 = pd.read_csv("../Results/bestProb_results/Schaffers3_prob.csv")


prob_bent_data = [df_prob_bent1['MBF'], df_prob_bent2['MBF'], df_prob_bent3['MBF'],bent_data[0]]
prob_schaf_data = [df_prob_schaf1['MBF'], df_prob_schaf2['MBF'][0:9], df_prob_schaf3['MBF'], schaf_data[0]]


sub_bent = plt.subplot(131)
plt.boxplot(prob_bent_data, medianprops=medianprops)
sub_bent.set_title('Bent Cigar')
sub_bent.set_xticklabels(['Scheme 1','Scheme 2','Scheme 3','Benchmark'])


sub_schaf = plt.subplot(132)
plt.boxplot(prob_schaf_data, medianprops=medianprops)
sub_schaf.set_xticklabels(['Scheme 1','Scheme 2','Scheme 3','Benchmark'])
sub_schaf.set_title('Schaffers F7')


plt.show()
```
